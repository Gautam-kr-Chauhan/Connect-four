```python
print("Enter size of the grid")
row=int(input(("Number of rows:")))
col=int(input(("Number of columns:")))
grid=[]
conn={}
for i in range(row):
    tempGrid=[]
    for j in range(col):
        tempGrid.append(j+i*col)
        conn[j+i*col]=[]
    grid.append(tempGrid)
#     conn.append([False]*col)
# conn[0].append(1)
# conn[0].append(4)
# print(grid)
# print(conn[0])
## for Disply
def dig(num):
    dig=0
    if num==0:
        return 1
    while num!=0:
        num//=10
        dig+=1
    
    return dig

def display(grid,conn):
    for i in range(row):
        rowstr=""
        for j in range(col):
            rowstr+=str(grid[i][j])
            if(j<col-1 and grid[i][j+1] in conn[j+i*col]):
                rowstr+="-"*(5-dig(grid[i][j]))
            else:
                rowstr+=" "*(5-dig(grid[i][j]))
        print(rowstr)
        colstr=""
        for j in range(grid[i][0],grid[i][col-1]+1):
            if i<row-1 and grid[i+1][j%col] in conn[j]:
                colstr+="|    "
            else:
                colstr+="     "
        print(colstr)
def connect(n1,n2):
    conn[n1].append(n2)
    display(grid,conn)
display(grid,conn)
## end of display task

##Box manipulation
# finding all possible boxes
allbox=[]
for i in range(row-1):
    for j in range(col-1):
        box=[]
        box.append((grid[i][j],grid[i][j+1]))
        box.append((grid[i][j],grid[i+1][j]))
        box.append((grid[i+1][j],grid[i+1][j+1]))
        box.append((grid[i][j+1],grid[i+1][j+1]))
        allbox.append(box)

        
def ispresent(n1,n2):
    for box in allbox:
        if (n1,n2) in box:
            return True
    return False
player1=input("Enter name of the player1:")
player2=input("Enter name of the player2:")
turn1=True
score1=score2=0
while True:
    if turn1:
        print(f"{player1} it is your turn")
        
        turn1=False
    else:
        print(f"{player2} it is your turn")
        turn1=True
    try:
        n1,n2=map(int,input("Enter two adjcent points (format: a b) to connect:").split())
    except ValueError:
        print("Point entered in wrong format!!")
        turn1=not turn1
        continue
        
    n1,n2=min(n1,n2),max(n1,n2)
    if not ispresent(n1,n2):
        print("Invalid connecting points")
        turn1=not turn1
        continue
    connect(n1,n2)
    for bx in allbox:
        if (n1,n2) in bx:
            bx.remove((n1,n2))
    count=allbox.count([])     
    if count>0:
        if not turn1:
            score1+=count
            turn1=True
        else:
            turn1=False
            score2+=count
    while(allbox.count([])>0):
        allbox.remove([])
    print(f"score1={score1},score2={score2}")
    #print(allbox)
    if allbox==[]:
        print("Game over!!")
        if score1>score2:
            print(f"{player1} is winner!!")
        elif score1<score2:
            print(f"{player2} is winner!!")
        else:
            print("Match tie")
        break
```
